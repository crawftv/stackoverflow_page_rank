```python
import pandas as pd
import numpy as np
import requests
import bs4
```

```python
df = pd.DataFrame()
df
```

```python
def get_page(i):
    res = requests.get("https://stackoverflow.com/questions/"+str(i))
    if res.status_code == requests.codes.ok:
        soup = bs4.BeautifulSoup(res.text)
        # get all links from answers from a post
        posts =  soup.find_all('div' , {'class':'post-text'})
        links = []
        for n in posts:
            a = n.find_all('a')
            links.append(a)
        flattened_list = [y for x in links for y in x]
        df[i] = [flattened_list]
    else: 
        pass
```
